```python
# import CMD
from berliner.parsec import CMD
# initialize CMD
c = CMD()
```

```python
# define your grid (logAge, [M/H])
grid_logage = (6, 10., 0.5)
grid_mh = (-2.5, 0.5, 0.5)
```

```python
# get isochrones
isoc_lgage, isoc_mhini, isoc_list_2mass_wise = c.get_isochrone_grid_mh(
    grid_logage=grid_logage, grid_mh=grid_mh, photsys_file="2mass_spitzer_wise",
    n_jobs=10, verbose=10)
```

```python
# plot age-metallicity
plot(isoc_lgage, isoc_mhini, 'x')
xlabel("log(Age/yr)")
ylabel("[M/H] (dex)");
```

```python
# plot isochrones
for isoc in isoc_list_2mass_wise:
    plot(isoc["logTe"], isoc["logg"])
xlim(5,3.3)
ylim(6,-2)
xlabel("log10(Teff/K)")
ylabel("logg (dex)");
```
